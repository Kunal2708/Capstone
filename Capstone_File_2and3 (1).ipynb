```python
for i in purl:
    driver.get(i)
    time.sleep(1)
    try:
        name = driver.find_element(By.XPATH,'//div[@class="h3U6cGyEUf76tvCpYisik"]/h1')
        name_address.append(name.text)
    except:
        name_address.append('-')
    try:
        pt = driver.find_element(By.XPATH,'//dl[@class="_4hBezflLdgDMdFtURKTWh"]/div[1]/dd/span/p')
        p_type.append(pt.text)
    except:
        p_type.append('-')    
    try:
        bed = driver.find_element(By.XPATH,'//dl[@class="_4hBezflLdgDMdFtURKTWh"]/div[2]/dd/span/p')
        bedroom.append(bed.text)
    except:
        bedroom.append('-')
    try:
        bath = driver.find_element(By.XPATH,'//dl[@class="_4hBezflLdgDMdFtURKTWh"]/div[3]/dd/span/p')
        bathroom.append(bath.text)
    except:
        bathroom.append('-')    
    try:
        price1 = driver.find_element(By.XPATH,'//div[@class="_1gfnqJ3Vtd1z40MlC0MzXu"]/span[1]')
        price.append(price1.text)
    except:
        price.append('-')    
    try:
        carea = driver.find_element(By.XPATH,'//dl[@class="_4hBezflLdgDMdFtURKTWh"]/div[4]/dd/span/p[1]')
        carpetarea.append(carea.text)
    except:
        carpetarea.append('-')
    try:
        gard = driver.find_element(By.XPATH,'//dl[@class="_1H0UoH91Rh0KKGH7JRGxEd"]/div[3]/dd/span')
        garden.append(gard.text)
    except:
        garden.append('-')
    try:
        stat = driver.find_element(By.XPATH,'//div[@class="_32IXcntQDSML3xVUkSdyXN"]')
        status.append(stat.text)
    except:
        status.append('-')
    try:
        park = driver.find_element(By.XPATH,'//dl[@class="_1H0UoH91Rh0KKGH7JRGxEd"]/div[2]/dd/span')
        parking.append(park.text)
    except:
        parking.append('-')
    try:
        tax= driver.find_element(By.XPATH,'//dl[@class="_1H0UoH91Rh0KKGH7JRGxEd"]/div[1]/dd')
        ctax.append(tax.text)
    except:
        ctax.append('-')
```
